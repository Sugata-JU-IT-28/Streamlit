```python
import re
import unicodedata

# 1. Load file
df = pd.read_csv("startup_funding.csv")

# 2. Column containing startup/company names
name_col = "Startup Name"

# 3. Common suffixes to remove
suffixes = [
    "technologies", "technology", "tech", "labs", "lab",
    "systems", "system", "solutions", "solution",
    "services", "service", "ventures", "venture",
    "foods", "food", "retail", "mobility",
    "cabs", "cab", "app", "online", "india",
    "pvt ltd", "private limited", "privatelimited",
    "ltd", "limited", "inc", "llp", "co", "company",
    "enterprise", "enterprises", "group", "digital"
]

# Convert suffixes into compact format
suffixes_compact = sorted(
    {re.sub(r"[^a-z0-9]", "", s.lower()) for s in suffixes},
    key=len,
    reverse=True
)

def clean_text(s):
    if pd.isna(s):
        return ""
    s = str(s)

    # convert fancy apostrophes etc. to normal ASCII
    s = unicodedata.normalize("NFKD", s).encode("ascii", "ignore").decode("ascii")

    # lowercase
    s = s.lower()

    # replace & with and
    s = s.replace("&", " and ")

    # remove apostrophes
    s = re.sub(r"[\"'`’]+", "", s)

    # keep only letters and numbers
    s = re.sub(r"[^a-z0-9]+", " ", s)

    # remove extra spaces
    s = re.sub(r"\s+", " ", s).strip()

    return s

def compact_text(s):
    return re.sub(r"[^a-z0-9]", "", clean_text(s))

def root_key(s):
    s = compact_text(s)

    # remove suffixes repeatedly
    changed = True
    while changed and s:
        changed = False
        for suf in suffixes_compact:
            if len(s) > len(suf) + 2 and s.endswith(suf):
                s = s[:-len(suf)]
                changed = True
                break
    return s

# 4. Create cleaned company name
df["Startup Name"] = df[name_col].apply(root_key)
df.head()


```

```python


usd_to_inr = 83.0

def usd_to_crore_rupees(x):
    if pd.isna(x):
        return np.nan

    s = str(x).strip().lower()
    s = s.replace("\u00a0", " ").replace("\xa0", " ")
    s = re.sub(r"(\\\\xc2\\\\xa0|\\xc2\\xa0)", " ", s)

    if s in {"", "nan", "none", "n/a", "na", "unknown", "undisclosed"}:
        return np.nan

    s = re.sub(r"[^0-9.,+]", "", s)
    s = s.replace("+", "")
    s = s.replace(",", "")

    try:
        usd = float(s)
        return round((usd * usd_to_inr) / 10000000, 2)
    except:
        return np.nan

df["Amount in USD"] = df["Amount in USD"].apply(usd_to_crore_rupees)

#print(df[["Amount in USD", "Amount in INR Cr"]].head(20))

#df.drop("Amount in USD",axis=1,inplace = True)
```
